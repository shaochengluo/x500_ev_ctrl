```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Display plots inline
%matplotlib inline
```

```python
# Path to your VICON CSV file
# Load the CSV file
df = pd.read_csv("/home/cpsl/px4_ws/attack3_log36/csv_out/vicon__x500_7__x500_7.csv")

# Show basic info
print("Data summary:")
df.info()

# Display first few rows
df.head()
```

```python
# Extract ENU position components
x_e = df["transform.translation.x"].to_numpy()  # East
y_e = df["transform.translation.y"].to_numpy()  # North
z_e = df["transform.translation.z"].to_numpy()  # Up

# Convert to NED
x_n = y_e              # North
y_n = x_e              # East
z_n = -z_e             # Down

# Add to DataFrame
df["pos_north"] = x_n
df["pos_east"]  = y_n
df["pos_down"]  = z_n

print("✅ Converted ENU → NED position.")
df[["pos_north", "pos_east", "pos_down"]].head()
```

```python

# ---- Quaternion components (from CSV) ----
qx = df["transform.rotation.x"].to_numpy()
qy = df["transform.rotation.y"].to_numpy()
qz = df["transform.rotation.z"].to_numpy()
qw = df["transform.rotation.w"].to_numpy()

# ---- Build ENU rotation matrix elements (row-major) ----
r00 = 1 - 2*(qy**2 + qz**2)
r01 = 2*(qx*qy - qz*qw)
r02 = 2*(qx*qz + qy*qw)

r10 = 2*(qx*qy + qz*qw)
r11 = 1 - 2*(qx**2 + qz**2)
r12 = 2*(qy*qz - qx*qw)

r20 = 2*(qx*qz - qy*qw)
r21 = 2*(qy*qz + qx*qw)
r22 = 1 - 2*(qx**2 + qy**2)

# ---- Convert ENU rotation to NED rotation (left-multiply by mapping) ----
# R_map = [[0,1,0],[1,0,0],[0,0,-1]]
# Row0_ned = Row1_enu, Row1_ned = Row0_enu, Row2_ned = -Row2_enu
n00, n01, n02 = r10, r11, r12
n10, n11, n12 = r00, r01, r02
n20, n21, n22 = -r20, -r21, -r22

# ---- STANDARD extraction (Z-Y-X: yaw(psi), pitch(theta), roll(phi)) ----
# pitch = asin( clamp(-R[2,0], -1, 1) )
# roll  = atan2(R[2,1], R[2,2])
# yaw   = atan2(R[1,0], R[0,0])
pitch_std = np.arcsin(np.clip(-n20, -1.0, 1.0))
roll_std  = np.arctan2(n21, n22)
yaw_std   = np.arctan2(n10, n00)

# Convert to degrees
df["roll_std_deg"]  = np.degrees(roll_std)
df["pitch_std_deg"] = np.degrees(pitch_std)
df["yaw_std_deg"]   = np.degrees(yaw_std)

# ---- ALTERNATE extraction (swap roll <-> yaw) ----
# Some users / datasets use different axis-order naming; this variant swaps roll and yaw:
# roll_alt  = atan2(R[1,0], R[0,0])   (previously used for yaw)
# pitch_alt = asin( clamp(-R[2,0]) )  (same)
# yaw_alt   = atan2(R[2,1], R[2,2])   (previously used for roll)
# NOTE: This is only an example alternate mapping — not standard Z-Y-X.
roll_alt  = np.arctan2(n10, n00)   # uses element that standard mapping used for yaw
pitch_alt = pitch_std             # same
yaw_alt   = np.arctan2(n21, n22)  # uses element that standard mapping used for roll

df["roll_alt_deg"]  = np.degrees(roll_alt)
df["pitch_alt_deg"] = np.degrees(pitch_alt)
df["yaw_alt_deg"]   = np.degrees(yaw_alt)

# ---- Print a small table to inspect differences ----
cols = ["t_rel",
        "roll_std_deg", "pitch_std_deg", "yaw_std_deg",
        "roll_alt_deg", "pitch_alt_deg", "yaw_alt_deg"]
print(df[cols].head(10).to_string(index=False))

# ---- Quick correlation check to see which mapping is distinct ----
corr_std = df[["roll_std_deg","pitch_std_deg","yaw_std_deg"]].corr()
corr_alt = df[["roll_alt_deg","pitch_alt_deg","yaw_alt_deg"]].corr()
print("\nCorrelation matrix (standard mapping):\n", np.round(corr_std,3))
print("\nCorrelation matrix (alternate mapping):\n", np.round(corr_alt,3))
```

```python
time_s = df["t_rel"].to_numpy()

plt.figure(figsize=(10, 4))
plt.plot(time_s, df["pos_north"], label="North (m)")
plt.plot(time_s, df["pos_east"],  label="East (m)")
plt.plot(time_s, df["pos_down"],  label="Down (m)")
plt.xlabel("Time (s)")
plt.ylabel("Position (m)")
plt.title("VICON Position (NED) vs Time")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 4))
plt.plot(time_s, df["roll_deg"],  label="Roll (°)")
plt.plot(time_s, df["pitch_deg"], label="Pitch (°)")
plt.plot(time_s, df["yaw_deg"],   label="Yaw (°)")
plt.xlabel("Time (s)")
plt.ylabel("Angle (°)")
plt.title("Euler Angles (NED) vs Time")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
yaw_unwrapped = np.unwrap(np.radians(df["yaw_deg"].to_numpy()))
plt.figure(figsize=(10, 4))
plt.plot(time_s, np.degrees(yaw_unwrapped), label="Unwrapped Yaw (°)")
plt.xlabel("Time (s)")
plt.ylabel("Angle (°)")
plt.title("Continuous Yaw vs Time")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
